Se detiene cuando llega a una decisiÃ³n final: clase (benigno/maligno) o valor (regresiÃ³n).

**Random Forest**

En un modelo de regresiÃ³n logÃ­stica, los coeficientes (coef_) indican quÃ© tanto influye cada variable en que el modelo prediga la clase 1 (por ejemplo, diagnÃ³stico = 1: Maligno).

El grÃ¡fico generalmente muestra:

En el eje Y â†’ las variables

En el eje X â†’ los coeficientes (importancia)

*Signo del coeficiente*

âž• Coeficiente positivo: La variable aumenta la probabilidad de que la clase sea 1 (Maligno).

âž– Coeficiente negativo: La variable disminuye la probabilidad de clase 1 (Maligno).


La magnitud indica quÃ© tan importante es la variable:

|coef| grande â†’ variable con mayor impacto en la predicciÃ³n

|coef| pequeÃ±o â†’ variable con poco efecto


**Â¿Por quÃ© utilizamos el modelo?** ðŸ’¡

La **RegresiÃ³n** logÃ­stica es un modelo de aprendizaje supervisado utilizado principalmente para clasificaciÃ³n binaria, es decir, cuando queremos decidir entre dos clases (0 o 1).

1. Empieza como una regresiÃ³n lineal

Primero, el modelo calcula una combinaciÃ³n lineal de las variables donde:

\begin{equation}
z= w_1 x_1 + w_2 x_2 +...+ w_n x_n + b
\end{equation}

donde:

$ð‘¥_i$: variable

$ð‘¤_i$: coeficiente aprendido

$b$: sesgo (intercepto)

2. Aplica la funciÃ³n sigmoide para convertir eso en una probabilidad

La regresiÃ³n lineal puede dar cualquier nÃºmero (â€“âˆž, +âˆž),
pero la sigmoide lo transforma a un valor entre 0 y 1:

\begin{equation}
\sigma(z)=\frac{1}{1+e^{-z}}
\end{equation}

Entonces el modelo predice:
- Probabilidad de clase 1 = $\sigma(z)$
- Probabilidad de clase 0 = 1 - $\sigma(z)$

3. Decide la clase usando un umbral

- Si probabilidad â‰¥ 0.5 â†’ Clase 1
- Si probabilidad < 0.5 â†’ Clase 0

El umbral puede ajustarse

4. Â¿CÃ³mo aprende el modelo?

encuentra los coeficientes ð‘¤ que mejor separan las clases, usando una funciÃ³n de pÃ©rdida llamada Log-Loss, que penaliza predicciones incorrectas.

- Calcula probabilidades
- Compara con los valores reales
- Ajusta los pesos para mejorar las predicciones

Este proceso se repite (iteraciones) hasta encontrar los mejores parÃ¡metros.

```python
# InstalaciÃ³n opcional (si faltara alguna librerÃ­a)
#!pip install pandas scikit-learn

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

# 1. Crear un pequeÃ±o dataset
data = {
    'texto': [
        "Estoy muy feliz hoy", "Me encanta este dÃ­a", "QuÃ© emociÃ³n verlas",
        "Estoy triste", "No quiero salir", "Estoy muy desanimada",
        "Estoy enojada", "QuÃ© coraje tengo", "No soporto esto"
    ],
    'emoji': ["feliz", "feliz", "feliz", "triste", "triste", "triste", "enojado", "enojado", "enojado"]
}

df = pd.DataFrame(data)

# 2. Dividir datos
X_train, X_test, y_train, y_test = train_test_split(df["texto"], df["emoji"], test_size=0.3, random_state=42)

# 3. Convertir texto a nÃºmeros
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# 4. Crear y entrenar el modelo
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# 5. Evaluar
y_pred = model.predict(X_test_vec)
print("Reporte de clasificaciÃ³n:\n")
print(classification_report(y_test, y_pred))

# 6. Probar con texto de la usuaria
while True:
    texto = input("Escribe un mensaje para predecir ('salir' para terminar): ")
    if texto.lower() == "salir":
        break
    texto_vec = vectorizer.transform([texto])
    pred = model.predict(texto_vec)[0]
    print("PredicciÃ³n:", pred)
```

```python
from datasets import load_dataset

# Ejemplo con EmotionSpanish
dataset = load_dataset("Cesar42/EmotionSpanish")
print(dataset["train"][0])
```

```python
import pandas as pd
import re

# Nombre de la columna que contiene todo el texto
raw_col = dataset["train"].column_names[0]

textos = []
labels = []

for item in dataset["train"][raw_col]:
    # Extraer label al final (nÃºmero)
    label_match = re.search(r'(\d+)\s*</s>', item)
    label = int(label_match.group(1)) if label_match else None
    labels.append(label)

    # Extraer la frase despuÃ©s de "Frase:" y antes de "[/INST]"
    text_match = re.search(r'Frase:\s*(.*?)\s*\[/INST\]', item)
    text = text_match.group(1) if text_match else item
    textos.append(text)

# Crear DataFrame
df = pd.DataFrame({
    "texto": textos,
    "emocion_num": labels
})

# Mapear emociones
emociones_map = {
    0: "alegrÃ­a",
    1: "miedo",
    2: "tristeza",
    3: "enojo",
    4: "asco",
    5: "sorpresa",
    6: "neutral"
}
df["emocion"] = df["emocion_num"].map(emociones_map)

df.head(10)
```

```python
# ðŸ“Œ Bloque completo: ClasificaciÃ³n de emociones en espaÃ±ol

# 1ï¸âƒ£ LibrerÃ­as
import pandas as pd
import re
from datasets import load_dataset
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix


# 5ï¸âƒ£ Dividir en train/test
X_train, X_test, y_train, y_test = train_test_split(df["texto"], df["emocion"], test_size=0.2, random_state=42)

# 6ï¸âƒ£ Vectorizar texto
vectorizer = TfidfVectorizer(max_features=2000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# 7ï¸âƒ£ Entrenar Logistic Regression
model = LogisticRegression(max_iter=500)
model.fit(X_train_vec, y_train)

# 8ï¸âƒ£ Evaluar
y_pred = model.predict(X_test_vec)
print("ðŸ“Œ Reporte de clasificaciÃ³n:\n")
print(classification_report(y_test, y_pred))
print("ðŸ“Œ Matriz de confusiÃ³n:\n")
print(confusion_matrix(y_test, y_pred))

# 9ï¸âƒ£ PredicciÃ³n interactiva
print("\nPrueba interactiva de emociones (escribe 'salir' para terminar)")
while True:
    texto = input("Escribe un mensaje: ")
    if texto.lower() == "salir":
        break
    texto_vec = vectorizer.transform([texto])
    pred = model.predict(texto_vec)[0]
    print("PredicciÃ³n:", pred)
```
